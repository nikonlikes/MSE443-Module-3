```python
import random
def generate_random_number(seed,begin,end):
    random.seed(seed)  # Set the random seed
    return random.randint(begin, end)
```

```python
import pandas as pd
import numpy as np
def saving(lst,name):
  df = pd.DataFrame(lst)

  # Saving the DataFrame to a CSV file
  df.to_csv(name, index=False, header=False)
  print(name)
  print(df)

  print("DataFrame saved to", name)
```

```python
import os

sample_configs = [
    {"folder": "small sized samples",  "orders": (6, 8),     "itemtypes": (7, 8), "totes": (6, 8)},
    {"folder": "medium sized samples", "orders": (30, 35),   "itemtypes": (7, 8), "totes": (16, 20)},
    {"folder": "large sized samples",  "orders": (250, 300), "itemtypes": (7, 8), "totes": (150, 200)},
]

n_datasets = 30

for config in sample_configs:
    output_dir = config["folder"]
    os.makedirs(output_dir, exist_ok=True)
    print(f"\n{'='*60}")
    print(f"  Generating {n_datasets} datasets in '{output_dir}'")
    print(f"{'='*60}")

    for dataset_idx in range(1, n_datasets + 1):
        seed = dataset_idx * 100
        random.seed(seed)

        n_orders = random.randint(*config["orders"])
        n_itemtypes = random.randint(*config["itemtypes"])
        n_totes = random.randint(*config["totes"])

        print(f"\n--- Dataset {dataset_idx} (seed={seed}) ---")
        print(f"  n_orders={n_orders}, n_itemtypes={n_itemtypes}, n_totes={n_totes}")

        # Generate order_itemtypes and order_quantities
        order_itemtypes = [[] for _ in range(n_orders)]
        order_quantities = [[] for _ in range(n_orders)]
        for i in range(n_orders):
            order_size = random.randint(1, 3)
            tt = random.sample(range(0, n_itemtypes - 1), order_size)
            qq = [random.randint(1, 3) for _ in range(order_size)]
            order_itemtypes[i] = tt
            order_quantities[i] = qq

        # Generate orders_totes
        orders_totes = [[] for _ in range(n_orders)]
        for i in range(n_orders):
            for j in range(len(order_itemtypes[i])):
                if j == 0:
                    orders_totes[i].append(random.randint(0, n_totes - 1))
                else:
                    if random.randint(0, 1) == 0:
                        orders_totes[i].append(orders_totes[i][0])
                    else:
                        orders_totes[i].append(random.randint(0, n_totes - 1))

        # Save with incrementing filenames into the output folder
        saving(order_itemtypes, os.path.join(output_dir, f"order_itemtypes_{dataset_idx}.csv"))
        saving(orders_totes, os.path.join(output_dir, f"orders_totes_{dataset_idx}.csv"))
        saving(order_quantities, os.path.join(output_dir, f"order_quantities_{dataset_idx}.csv"))
```
